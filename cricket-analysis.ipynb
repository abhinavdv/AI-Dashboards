```python
# azure_openai.py
import requests

def call_azure_openai(prompt):
    url = "https://<your-endpoint>.openai.azure.com/openai/deployments/<your-deployment>/completions?api-version=2023-05-15"
    headers = {
        "Content-Type": "application/json",
        "api-key": "<your-api-key>"
    }
    data = {
        "prompt": prompt,
        "max_tokens": 100,
        "temperature": 0,
        "model": "gpt-40"
    }
    
    response = requests.post(url, headers=headers, json=data)
    return response.json()
```
